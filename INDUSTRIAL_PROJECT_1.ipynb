# ```Checkpoint 6: Plots``` 

````python
import matplotlib.pyplot as plt
import seaborn as sb
````

````python
fi=final_master_frame.groupby(by='funding_round_type')['raised_amount_usd'].mean().sort_values(ascending=False).astype(int)
funding_type_lst = ['venture','seed','private_equity',]
Average_amt=[fi[1],fi[3],fi[0]]
````

````python
fj=final_master_frame.groupby(by='funding_round_type')['raised_amount_usd'].sum().sort_values(ascending=False)
funding_type_lst = ['venture','seed','private_equity']
total_amt=[fj[0],fj[2],fj[1]]
````

````python
plt.figure(figsize=(15,8))

plt.subplot(121)
plt.yscale('log')
#plt.grid()
sb.set_style(style="darkgrid")
plt.title("Average Amount Of Investment",fontsize=18)
plt.xlabel("Funding Types",fontsize=18)
plt.ylabel("Raised Amount Usd",fontsize=18)
#plt.yscale('log')
plt.bar(funding_type_lst,Average_amt,color="blue",label="Average_amount")
plt.legend()


plt.subplot(122)
plt.yscale('log')
#plt.grid()
sb.set_style(style="darkgrid")
plt.title("Total Amount Of Investment",fontsize=18)
plt.xlabel("Funding Types",fontsize=18)
plt.ylabel("Raised Amount Usd",fontsize=18)
plt.bar(funding_type_lst,total_amt,color="purple",label="Total_amount")
plt.legend()
````
